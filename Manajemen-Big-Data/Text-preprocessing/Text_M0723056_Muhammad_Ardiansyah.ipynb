# Preprocessing

```python
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# ubah teks ke lower
News['text'] = News['text'].str.lower()
News
```

```python
# cek apakah text memiliki link
def find_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.findall(text)

News_cek = News['text'].apply(find_urls)
print("Number of rows with URLs:", News[News_cek.apply(len) > 0].shape[0])
```

```python
# Hapus link
News["text"] = News["text"].apply(lambda x: re.sub(r'http\S+|www\S+|https\S+', '', x))  

# Hapus Angka
News["text"] = News["text"].apply(lambda x: re.sub(r'\d+', '', x))

# Hapus tanda baca
News["text"] = News["text"].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))

# Hapus stopwords
stop_words = set(stopwords.words('english'))
News["text"] = News["text"].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python
News
```

# Tokenization

```python
#Tokenization
from nltk.tokenize import word_tokenize
news_token = News['text'].apply(word_tokenize)

print(news_token)
```

Tokenization hanya memecah teks menjadi token (kata per kata) namun bentuk kata tidak di normalisasikan maka masih banyak noise dalam bentuk variasi kata yang seharusnya sama.

## Lemmatization

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()
News_lem= News["text"].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
```

```python
print(News_lem)
```

Lemmatization mengubah bentuk kata menjadi dasar sehingga ini akan memperkecil jumlah fitur, karena tujuan dari ini untuk klasifikasi apakah berita ini fake/true maka lemmatization lebih baik daripada tokenization.

## TF_IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
news_tfidf = tfidf_vectorizer.fit_transform(News_lem)
```

```python
# Bentuk matriks TF-IDF:
print("Bentuk matriks TF-IDF:", news_tfidf.shape)
# Jumlah entri bukan nol :
print("Jumlah entri bukan nol (nnz):", news_tfidf.nnz)
total_cells = news_tfidf.shape[0] * news_tfidf.shape[1]
sparsity = 1.0 - news_tfidf.nnz / total_cells
print(f"Sparsitas: {sparsity:.6f}")

# tampilkan jumlah fitur dan 50 nama fitur pertama
feature_names = tfidf_vectorizer.get_feature_names_out()
print("Jumlah total fitur:", feature_names.shape[0])
print("50 nama fitur pertama:", feature_names[:50])
```

## Word Embedding

```python
from gensim.models import Word2Vec
import warnings
warnings.filterwarnings('ignore')

# Tokenisasi teks yang telah di lemmatization
tokenized_news = [news.split() for news in News_lem]

# Melatih model Word2Vec
# Parameters:
# - sentences: data teks yang sudah ditokenisasi.
# - vector_size: dimensi dari vektor kata yang akan dihasilkan (misal: 100).
# - window: jarak maksimum antara kata saat ini dan kata yang diprediksi dalam satu kalimat.
# - min_count: mengabaikan semua kata dengan frekuensi total lebih rendah dari ini.
# - workers: jumlah thread worker untuk melatih model.

word2vec_model = Word2Vec(sentences=tokenized_news, vector_size=100, window=5, min_count=1, workers=4)
```

```python
# Menampilkan informasi tentang model Word2Vec yang telah dilatih
print("Objek Word2Vec KeyedVectors (word2vec_model.wv):")
print(word2vec_model.wv)
vocab_size = len(word2vec_model.wv.key_to_index)
vector_dimension = word2vec_model.wv.vector_size
print(f"Bentuk Word2Vec (kosakata, dimensi vektor): ({vocab_size}, {vector_dimension})")

# Tampilkan beberapa contoh saja jika kosakatanya sangat besar
print("Nama fitur (kosakata Word2Vec):")
feature_names_w2v = list(word2vec_model.wv.key_to_index.keys())
print(feature_names_w2v[:50]) # Menampilkan 50 kata pertama
if len(feature_names_w2v) > 50:
    print(f"... dan {len(feature_names_w2v) - 50} kata lainnya.")
```

word embedding dengan word2vec menghasilkan dimensi yang jauh lebih kecil jika dibandingkan tf-idf
